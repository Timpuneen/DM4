```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("creditcard.csv")
```

```python
print(df.info())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Amount'], kde=True, bins=30)
plt.title('Distribution of Transaction Amount')
plt.subplot(1, 2, 2)
sns.histplot(df['Time'], kde=True, bins=30)
plt.title('Distribution of Transaction Time')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), cmap="coolwarm", annot=False, fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
#Timur Baltabayev
#2. Data Preprocessing
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.fillna(df.mean(), inplace=True)
```

```python
scaler = StandardScaler()
df[['Amount', 'Time']] = scaler.fit_transform(df[['Amount', 'Time']])
```

```python
#Timur Baltabayev
#3. Anomaly Detection Techniques
```

```python
from scipy.stats import zscore
```

```python
#Z-score
df['Z_score'] = zscore(df['Amount'])
df['Anomaly_Z'] = df['Z_score'].apply(lambda x: 1 if abs(x) > 3 else 0)
```

```python
from sklearn.ensemble import IsolationForest
```

```python
#Isolation Forest
iso_forest = IsolationForest(contamination=0.01)
df['Anomaly_IF'] = iso_forest.fit_predict(df.drop(['Class'], axis=1))
```

```python
df['Anomaly_IF'] = df['Anomaly_IF'].apply(lambda x: 1 if x == -1 else 0)
```

```python
#Timur Baltabayev
#4. Model Evaluation
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import numpy as np
```

```python
y_true = df['Class']
y_pred_if = df['Anomaly_IF']
```

```python
print("Isolation Forest:")
print(classification_report(y_true, y_pred_if))
print(confusion_matrix(y_true, y_pred_if))
```

```python
#Timur Baltabayev
#5. Visualization
```

```python
sns.heatmap(confusion_matrix(y_true, y_pred_if), annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - Isolation Forest")
plt.show()
```

```python
#Curve for Isolation Forest
fpr_if, tpr_if, _ = roc_curve(y_true, y_pred_if)
roc_auc_if = auc(fpr_if, tpr_if)
```

```python
plt.figure()
plt.plot(fpr_if, tpr_if, color='darkorange', lw=2, label=f'Isolation Forest (area = {roc_auc_if:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic - Isolation Forest')
plt.legend(loc="lower right")
plt.show()
```

```python
import pandas as pd

df = pd.read_csv('storms.csv')

df.info()
df.describe()
df.head()
```

```python
#Timur Baltabayev
#2. Data Preprocessing
```

```python
df = df.ffill()  
```

```python
#'year', 'month', 'day', 'hour' to a datetime format
df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
```

```python
#datetime as the index
df.set_index('datetime', inplace=True)
```

```python
df['wind'] = pd.to_numeric(df['wind'], errors='coerce') 
df['wind'] = df['wind'].ffill()  
df_resampled = df[['wind']].resample('D').mean()#   'wind'
```

```python
#Timur Baltabayev
#3. Exploratory Data Analysis
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df_resampled['wind'], label='Wind Speed')
plt.title('Hurricane Wind Speed Over Time')
plt.xlabel('Date')
plt.ylabel('Wind Speed (knots)')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

df_resampled['wind'] = df_resampled['wind'].ffill()
# Decompose the time series
result = seasonal_decompose(df_resampled['wind'], model='additive', period=365)
result.plot()
plt.show()
```

```python
#Timur Baltabayev
#4. Modeling
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(df_resampled['wind'], order=(5, 1, 0))
model_fit = model.fit()

#predictions
predictions = model_fit.forecast(steps=30)  # Forecast next 30 days
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

#Holt-Winters model
model_hw = ExponentialSmoothing(df_resampled['wind'], trend='add', seasonal='add', seasonal_periods=365)
model_hw_fit = model_hw.fit()

#Forecast
forecast_hw = model_hw_fit.forecast(steps=30)
```

```python
from sklearn.ensemble import RandomForestRegressor

X = df_resampled['wind'].shift(1).dropna().values.reshape(-1, 1)
y = df_resampled['wind'].dropna().values
#---------------------------------------
if len(X) > len(y):
    X = X[:len(y)]
elif len(y) > len(X):
    y = y[:len(X)]
#---------------------------------------

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

#Train RF
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)

rf_predictions = rf_model.predict(X_test)
```

```python
#Timur Baltabayev
#5. Model Evaluation
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Calculate MAE, RMSE
mae = mean_absolute_error(y_test, rf_predictions)
rmse = np.sqrt(mean_squared_error(y_test, rf_predictions))
mape = np.mean(np.abs((y_test - rf_predictions) / y_test)) * 100

print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
print(f'MAPE: {mape}')
```

```python
#Timur Baltabayev
#6. Visualize Predictions vs Actual
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y_test, label='Actual')
plt.plot(rf_predictions, label='Predicted')
plt.title('Predicted vs Actual Wind Speed')
plt.xlabel('Date')
plt.ylabel('Wind Speed (knots)')
plt.legend()
plt.show()
```
